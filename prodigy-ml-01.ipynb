```python
test_data=pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
test_data
```

```python
# Features that are related to the square footage and the number of bedrooms and bathrooms in house
features=['LotArea','1stFlrSF','2ndFlrSF',
          'TotalBsmtSF','BsmtFinSF2','BsmtFinSF1','BsmtUnfSF',
          'BedroomAbvGr',
          'BsmtFullBath','BsmtHalfBath','FullBath'
         ]

def prepare_data1(df): 
    data=df[features]
    data['Target']=train_data[train_data.columns[-1]]
    data.dropna(subset=features,inplace=True)  
    X_train=data[features[:-1]];y_train=data['Target']
    return X_train,y_train
```

```python
X_train,y_train=prepare_data1(train_data)
X_test,y_test=prepare_data1(test_data)
```

# Just trying the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler


model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_train)
mse = mean_squared_error(y_train, y_pred)

y_pred2=model.predict(X_test)
mse2 = mean_squared_error(y_test, y_pred2)

print(f'Mean Squared Error(train): {mse}')
print(f'Mean Squared Error(test): {mse2}')
```

# Scaled the columns mannually by **dividing with appropriate number**

```python
def prepare_data2(df): 
    data=df[features]
    data['Target']=train_data[train_data.columns[-1]]
    data.dropna(subset=features,inplace=True)  
    X=data[features[:-1]];y=data['Target']
    X['LotArea']=X['LotArea']/1000
    X['1stFlrSF']=X['1stFlrSF']/100
    X['2ndFlrSF']=X['2ndFlrSF']/100
    X['TotalBsmtSF']=X['TotalBsmtSF']/100
    X['BsmtUnfSF']=X['BsmtUnfSF']/100
    X['BsmtFinSF1']=X['BsmtFinSF1']/100
    X['BsmtFinSF2']=X['BsmtFinSF2']/100
    return X,y

X_train,y_train=prepare_data2(train_data)
X_test,y_test=prepare_data2(test_data)

model2 = LinearRegression()
model2.fit(X_train,y_train)

y_pred=model2.predict(X_train)
mse = mean_squared_error(y_train, y_pred)

y_pred2=model2.predict(X_test)
mse2 = mean_squared_error(y_test, y_pred2)
print(f'Mean Squared Error(train): {mse}')
print(f'Mean Squared Error(test): {mse2}')



```

# Using **Min-Max regularization** (Same min-max range for all the columns)

```python
def prepare_data3(df): 
    data=df[features+['Id']]
    data['Target']=train_data[train_data.columns[-1]]
    data.dropna(subset=features,inplace=True)  
    X=data[features[:-1]];y=data['Target']
    scaler = MinMaxScaler(feature_range=(0, 10))  # specify your own range here
    scaler.fit(X)
    X= scaler.transform(X)
    y=y/1000
    
    return X,y,data

X_train,y_train,data=prepare_data3(train_data)
X_test,y_test,data=prepare_data3(test_data)

model3 = LinearRegression()
model3.fit(X_train,y_train)

y_pred=model3.predict(X_train)
mse = mean_squared_error(y_train, y_pred)

y_pred2=model3.predict(X_test)
mse2 = mean_squared_error(y_test, y_pred2)
print(f'Mean Squared Error(train): {mse}')
print(f'Mean Squared Error(test): {mse2}')



```

# **Min-Max regularization** but different range for different columns

```python
def min_max_regularize(df,C,a=0,b=1):
    if C in df.columns:
        df[C] = a + ((df[C] - df[C].min()) * (b - a)) / (df[C].max() - df[C].min())
    return df


X_train,y_train=prepare_data1(train_data)
X_test,y_test=prepare_data1(test_data)

X_train=min_max_regularize(X_train,'LotArea',0,10)
X_test=min_max_regularize(X_test,'LotArea',0,10)

X_train=min_max_regularize(X_train,'1stFlrSF',0,10)
X_test=min_max_regularize(X_test,'1stFlrSF',0,10)
X_train=min_max_regularize(X_train,'2ndFlrSF',0,10)
X_test=min_max_regularize(X_test,'2ndFlrSF',0,10)

#  	 	 	
X_train=min_max_regularize(X_train,'TotalBsmtSF',0,10)
X_test=min_max_regularize(X_test,'TotalBsmtSF',0,10)

X_train=min_max_regularize(X_train,'BsmtFinSF2',0,10)
X_test=min_max_regularize(X_test,'BsmtFinSF2',0,10)

X_train=min_max_regularize(X_train,'BsmtFinSF1',0,10)
X_test=min_max_regularize(X_test,'BsmtFinSF1',0,10)

X_train=min_max_regularize(X_train,'BsmtUnfSF',0,10)
X_test=min_max_regularize(X_test,'BsmtUnfSF',0,10)

y_train=np.log(y_train)
y_test=np.log(y_test)
```

```python
model4 = LinearRegression()
model4.fit(X_train,y_train)

y_pred=model4.predict(X_train)
mse = mean_squared_error(y_train, y_pred)

y_pred2=model4.predict(X_test)
mse2 = mean_squared_error(y_test, y_pred2)
print(f'Mean Squared Error(train): {mse}')
print(f'Mean Squared Error(test): {mse2}')
```
